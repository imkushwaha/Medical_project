```python
import cv2
from PIL import Image
```

```python
img = cv2.imread("clear_image.jpg", flags=cv2.IMREAD_GRAYSCALE)
```

```python
img
```

```python
Image.fromarray(img).show()
```

```python
_, new_img = cv2.threshold(img,150,255,cv2.THRESH_BINARY)
```

```python
Image.fromarray(new_img).show()
```

```python
img = cv2.imread("dark_image.jpg", flags=cv2.IMREAD_GRAYSCALE)
img
```

```python
Image.fromarray(img).show()
```

```python
_, new_img = cv2.threshold(img,150,255,cv2.THRESH_BINARY)
```

```python
Image.fromarray(new_img).show()
```

#### Simple Thresholding is not working here,

### We will use Adaptive Thresholding

```python
new_image = cv2.adaptiveThreshold(
    img, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    61, 
    11)  
## 61 is block size and 11 is constant came after heat and trial method
Image.fromarray(new_image).show()
```

- This time ADAPTIVE THRESHOLDING worked well for darker image
- All those block value and constant value in argument i have used is came after hit and trial method, that will vary with different situation
